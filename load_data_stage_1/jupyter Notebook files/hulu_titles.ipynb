```python
import pandas as pd
from sqlalchemy import create_engine, text
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
POSTGRES_URI = os.getenv('POSTGRES_URI')
engine = create_engine(POSTGRES_URI)
```

```python
df = pd.read_csv("../csv_files/hulu_titles.csv")
```

```python
# creating the table if not exists
with engine.begin() as conn:
    query = text(
        """
            CREATE TABLE IF NOT EXISTS hulu_titles(
                show_id VARCHAR(20) PRIMARY KEY NOT NULL,
                type TEXT NOT NULL,
                title TEXT NOT NULL,
                director TEXT,
                "cast" TEXT,
                country TEXT,
                date_added TEXT, 
                release_year INT,
                rating TEXT,
                duration TEXT,
                listed_in TEXT, 
                description TEXT
            )
        """
    )
    conn.execute(query)

# truncating the table
with engine.begin() as conn:
    query = text("TRUNCATE TABLE hulu_titles")

    conn.execute(query)
```

```python
with engine.begin() as conn:
    df.to_sql(
        con=conn,
        name='hulu_titles',
        if_exists='append',
        index=False
    )
```
